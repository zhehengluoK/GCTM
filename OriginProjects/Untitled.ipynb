```python
import matplotlib.pyplot as plt
from brokenaxes import brokenaxes
%matplotlib inline
```

```python
x = [0.5, 1, 1.64872, 2, 3, 5]
_20ng = [0.37594, 0.37994, 0.38171, 0.37663, 0.38163, 0.37747]
imdb = [0.2002, 0.19619, 0.20401, 0.19958, 0.20037]
nips = [0.40891, 0.41902, 0.41891, 0.41618, 0.41306]
```

```python
_20ng = [_ - (0.38171-0.379) for _ in _20ng]
imdb = [_ - (0.20401-0.202) for _ in imdb]
nips = [_ - (0.41891-0.410) for _ in nips]
```

```python
plt.figure(dpi=1200) 
bax = brokenaxes(ylims=((0.18, 0.22), (0.365, 0.42)), hspace=0.06)
bax.plot(x[:5], _20ng[:5], 'black', marker='s', label='20NG')
bax.plot(x[:5], imdb, 'r', marker='o', label='IMDB')
bax.plot(x[:5], nips, 'b', marker='^', label='NIPS')
bax.legend()
bax.set_xticks([0.5,1.0,1.5,2.0,2.5,3.0,3.8])
bax.set_xlabel('weight of negative sampling', labelpad=25)
bax.set_ylabel('NPMI')
plt.savefig('weight of negtive sampling.pdf')
plt.show()
```
